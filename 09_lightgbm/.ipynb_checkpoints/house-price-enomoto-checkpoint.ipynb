## 4. データの分析

```python
# 前処理ができたところでデータ分析に入ります。

train_df = df_mod
```

```python
corrmat = train_df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, cmap='Reds',vmax=.8, square=True); 
```

```python
# 販売価格との相関行列（ヒートマップ）
k =15 # 変数の数
# 関数の意味を記述する nlargest 大きい順にいくつかとる <->nsmallest 
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
# 販売価格との相関係数が大きい順に10こ並べて、そのインデックスをcolsに代入　
cm = np.corrcoef(train_df[cols].values.T)# 販売価格との相関係数が大きい順１０項目の値で再度相関係数をとる
# なぜ転置をとるのか http://lofas.hatenablog.com/entry/2015/02/09/150552
sns.set(font_scale=1.25) # 項目の名前のフォントを1.25倍する
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```

## 6.モデル作成

いよいろモデル学習と評価に入ります。
データ分析から得られた販売価格に影響しそうな10個の項目を使ってモデル学習を行います。

さて準備はできた。たくさん予測モデルアルゴリズムが選択肢としてある。 問題のタイプや要件を理解して、ここから選定していく。 
この問題は回帰でトレーニングセットが、与えられたモデルから予測するので教師あり学習になります。 これらの２つの観点（教師あり学習、分類と回帰）から絞ります。

- K近傍法

- パーセプトロン

つぎにデータを用意します。

```python
# 目的変数以外をトレーニングデータとする
X = train_df[cols]
# SalePriceを目的変数とする
Y = train_df['SalePrice']
# テストデータの必要な項目
#X_test = test_df[cols]
# 項目数がテストとトレーニングセットで同じか確認する
#X_train.shape, Y_train.shape ,X_test.shape
```

```python
# 機械学習アルゴリズムモジュールをインポート

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dict = {
    "LogisticRegression": LogisticRegression(),
    "SVM" : SVC(),
    "KNN" : KNeighborsClassifier(n_neighbors = 3),
    "Gaussian" : GaussianNB(),
    'Perceptron' : Perceptron(),
    'Linear SVC'  : LinearSVC(),
    'Stochastic Gradient Descent' : SGDClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest' : RandomForestClassifier(n_estimators=100),
}
```

```python
len(model_dict)
```

```python
import sklearn.cross_validation as cross_validation

#n_fold = 10 # 交差検定の回数
#k_fold = cross_validation.KFold(n=len(X_train),n_folds = n_fold)

# score = np.zeros(len(model_dict), 2)
Y_pred={}

for model_name, model_instance in model_dict.items():
    X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=0.2, random_state=0)
    print(model_name + "での学習" )
    # for train, test in k_fold:
    model_instance.fit(X_train, Y_train)
    Y_pred[model_name] = model_instance.predict(X_test)
    print(round(model_instance.score(X_train, Y_train) * 100, 2)) # 決定係数を小数点第２位まで%表示
    #print(round(model_instance.score(X_test, y_test) * 100, 2)) # 決定係数を小数点第２位まで%表示
    
```

## 3.前処理

- 欠損値の対処

```python
# 機械学習アルゴリズムモジュールをインポート

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dict = {
    "LogisticRegression": LogisticRegression(),
    "SVM" : SVC(),
    "KNN" : KNeighborsClassifier(n_neighbors = 3),
    "Gaussian" : GaussianNB(),
    'Perceptron' : Perceptron(),
    'Linear SVC'  : LinearSVC(),
    'Stochastic Gradient Descent' : SGDClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest' : RandomForestClassifier(n_estimators=100),
}
```

```python
len(model_dict)
```

```python
# RMSE
import math
def root_mean_square_error(y_test, y_pred):
   sum_rmse = 0
   for pred_price, test_price in zip(y_pred, y_test):
       sum_rmse += (pred_price - test_price) ** 2
   sum_rmse /= len(y_pred)
   sum_rmse =math.sqrt(sum_rmse)
   return sum_rmse
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import sklearn.cross_validation as cross_validation

#n_fold = 10 # 交差検定の回数
#k_fold = cross_validation.KFold(n=len(X_train),n_folds = n_fold)

# score = np.zeros(len(model_dict), 2)
best_rmse = 1000000000000

best_Y_pred = None
best_model = ""

for model_name, model_instance in model_dict.items():
    X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X, Y, test_size=0.2, random_state=0)
    print(model_name + "での学習" )
    # for train, test in k_fold:
    model_instance.fit(X_train, Y_train)
    Y_pred = model_instance.predict(X_test2)
    Y_pred_valid = model_instance.predict(X_test)
    #rmse = round(root_mean_square_error(Y_test, Y_pred_valid), 2)
    rmse = np.sqrt(mean_squared_error(list(Y_test), list(Y_pred_valid)))
    print(rmse) # 決定係数を小数点第２位まで%表示
    if rmse < best_rmse:
        best_rmse = rmse
        best_Y_pred = Y_pred
        best_model = model_name

print("ベストモデルは" + best_model )
print(best_rmse)
```
